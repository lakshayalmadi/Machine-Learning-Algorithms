```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Ads_CTR_Optimisation.csv')
df.head()
```

```python
#random selection of the algorithm
import random
N=10000
d=10
ads_selected=[]
total_reward=0
for i in range(0,N):
    ad=random.randrange(d)
    ads_selected.append(ad)
    reward=df.values[i,ad]
    total_reward=total_reward+reward
total_reward
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```

```python
#implementing UCB from scratch
numbers_of_selections=[0]*d
sums_of_rewards=[0]*d
```

```python
import math
ads_selected=[]

for i in range(0,N):
    max_upper_bound=0
    ad=0
    for j in range(0,d):
        if(numbers_of_selections[j]>0):
            average_reward=sums_of_rewards[j]/numbers_of_selections[j]
            delta_j=math.sqrt(3/2*math.log(i+1)/numbers_of_selections[j])
            upper_bound=average_reward+delta_j
        else:
            upper_bound=1e400
        if upper_bound>max_upper_bound:
            max_upper_bound=upper_bound
            ad=j
        
    ads_selected.append(ad)
    numbers_of_selections[ad]+=1
    reward=df.values[i, ad]
    sums_of_rewards[ad]+=reward
    total_reward+=reward    
```
